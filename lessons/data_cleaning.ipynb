```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams["figure.figsize"] = (20,10)
```

```python
df1 = pd.read_csv("data_sets/bengaluru_house_prices.csv")
df1.head()
```

```python
df1.shape
```

```python
df1.groupby('area_type')['area_type'].agg('count')
```

```python
df2 = df1.drop(['area_type', 'society', 'balcony', 'availability'], axis='columns')
df2.head()
```

```python
df2.isnull().sum()
```

```python
# drop null values
df3 = df2.dropna()
df3.isnull().sum()
```

```python
df3['size'].unique()
```

```python
# extract number of bedrooms from 'size' column
def extract_bhk(x):
    try:
        return int(x.split(' ')[0])
    except:
        return None
```

```python
df3["bhk"] = df3['size'].apply(extract_bhk)
df3["bhk"].unique()
```

```python
df3.head()
```

```python
df3[df3.bhk>20]
```

```python
df3.total_sqft.unique()
```

```python
# clean total_sqft column
def convert_sqft_to_num(x):
    try:
        tokens = x.split('-')
        if len(tokens) == 2:
            return (float(tokens[0]) + float(tokens[1])) / 2
        try:
            return float(x)
        except:
            return None
    except:
        return None
```

```python
df3["total_sqft"] = df3["total_sqft"].apply(convert_sqft_to_num)
df3.total_sqft.unique()
```
